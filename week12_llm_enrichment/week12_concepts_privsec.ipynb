### ðŸ§  Concept: The Reasoning Engine (Code vs AI)

| Task | Regular Code (Python) | AI (LLM) |
| :--- | :--- | :--- |
| **Logic** | Exact Rules (`if "cookie" in text`) | Fuzzy Logic (`Does this talk about cookies?`) |
| **Best For** | Counting, Math, Scraping | Summarizing, Classifying, reasoning |
| **Cost** | Free & Fast | Costs money/credits & Slower |

**The Pattern**: Use Python to *get* the data. Use AI to *understand* it.

````python
# Provider-agnostic interface (you can adapt for Gemini/OpenAI/etc.)
def llm_label_section(text: str) -> dict:
    """Return a JSON-like dict with labels and a short summary.
    
    Replace the body with a real API call in the assignment notebook.
    """
    # Placeholder heuristic for now:
    labels = {
        "choices_controls": bool(re.search(r"\b(opt\s?-?out|preferences|your choices|controls?)\b", text, re.I)),
        "retention": bool(re.search(r"\b(retention|retain)\b", text, re.I)),
        "third_party": bool(re.search(r"\b(third\s?-?party|share|sharing)\b", text, re.I)),
        "security": bool(re.search(r"\b(encrypt|encryption|security|safeguards)\b", text, re.I)),
    }
    return {"labels": labels, "summary": text[:200] + ("..." if len(text) > 200 else "")}
````
